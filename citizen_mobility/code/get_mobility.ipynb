```python
#筛选需要的字段
select_cols = ['person_id', 'person_num', 'day_num','trip_num', 'trip_duration', 'trip_distance', 'o_purpose', 'o_purpose_category', 'd_purpose', 'd_purpose_category', 'num_non_hh_travelers', 'num_hh_travelers',
               "num_travelers", "mode_type", "driver", "park_location", "park_pay"]
select_df_data = df[select_cols]
```

```python
driver = select_df_data.pop('driver')
```

```python
import tensorflow as tf
dataset = tf.data.Dataset.from_tensor_slices((select_df_data.values, driver.values))
```

```python
for feat, targ in dataset.take(5):
      print ('Features: {}, Target: {}'.format(feat, targ))
```

```python
train_dataset = dataset.shuffle(len(select_df_data)).batch(1)
```

```python
def get_compiled_model():
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(10, activation='relu'),
        tf.keras.layers.Dense(10, activation='relu'),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam',
                loss='binary_crossentropy',
                metrics=['accuracy'])
    return model
```

```python
#LSTM
def get_compiled_LSTMmodel():
    model=tf.keras.models.Sequential()
    model.add(tf.keras.layers.Embedding(output_dim=32,
                                        input_dim=16,
                                        input_length=28))
    #LSTM
    #全连接层
    model.add(tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(units=8)))
    model.add(tf.keras.layers.Dense(units=32,activation='relu'))
    model.add(tf.keras.layers.Dropout(0.3))

    #LSTM
    #输出层
    model.add(tf.keras.layers.Dense(units=2, activation='softmax'))
    model.summary()
    return model
```

```python
model = get_compiled_model()
```

```python
LSTMmodel=get_compiled_LSTMmodel()
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(0.003),
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(train_dataset, epochs=10)
```

```python
#训练过程可视化
import matplotlib as plt


def visu_train_history(train_history, train_metric, validation_metric):
    plt.plot(train_history.history[train_metric])
    plt.plot(train_history.history[validation_metric])
    plt.title('Train History')
    plt.ylabel('train_metric')
    plt.xlabel('epoch')
    plt.legend(['train', 'validation'], loc='upper left')
    plt.show()
```
